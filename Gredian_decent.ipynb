```python
# gredian decent for linear regression
# yhat = mx + c
# loss = ((y - yhat)^2)/N
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.random.randn(10,1)
y = 2*x +np.random.randn()
```

```python
# parameters
m = 0
c = 0
# hperparameter (learning rate)
learning_rate = 0.01
```

```python
def decent(x,y,m,c,learning_rate):
  dldm = 0
  dldc = 0
  N = x.shape[0]
  # iterate the loss fun
  # loss = ((y - (mx+c))^2)/N

  for xi, yi in zip(x,y):
    dldm += -2*xi*(yi-(m*xi+c))
    dldc += -2*(yi-(m*xi+c))


  m = m - learning_rate*(1/N)*dldm
  c = c - learning_rate*(1/N)*dldc

  return m,c

for epoch in range(400):
  m , c = decent(x,y,m,c,learning_rate)
  yhat = m*x + c
  loss = np.divide(np.sum((y-yhat)**2,axis = 0),x.shape[0])
  print(f'{epoch} : loss is {loss}, parameter m is {m}, parameter c is {c}')
  plt.plot(x,yhat)



```
